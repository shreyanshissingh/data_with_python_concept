### GET Requests

```python
import requests
import json

url = "https://jsonplaceholder.typicode.com/posts/12"
response = requests.get(url)
data = response.json()
data_1 = json.dumps(data, indent=4)
print(data)
print(data_1)

```

### POST

```python
url = "https://jsonplaceholder.typicode.com/posts"
data = {
    "userId": 12,
    "title": "Shreyansh Post",
    "body": "This is a post created by Shreyansh"
}

response = requests.post(url, json=data)
print(response.status_code)
print(response.json())
```

### Query Parameters

```python
url = "https://jsonplaceholder.typicode.com/posts"

params = {
    "userId": 1
}

response = requests.get(url, params=params)
print(response.status_code)
data = json.dumps(response.json(), indent=4)
print(data)
```
