```python
from pydantic import BaseModel
```

```python
class EmployeeSchema(BaseModel):
    name:str
    id:int
    dept:str
```

```python
obj = EmployeeSchema(name="Shre", id=1 , dept="HR")
obj_1 = EmployeeSchema(**{"name":"Shreyansh", "id":1 , "dept":"HR_12"})

print(f"Mehthod-1:\n {obj}")
print(f"Mehthod-2:\n {obj_1}")
```

```python
obj_3 = EmployeeSchema(name="Shre", id="1a" , dept="111")

print(obj_3)
```

If Pydantic can convert it would , here ID is string but it will accept the value if it can convert that sttring to ID<br>
This is called Coercion

```python
obj_4 = EmployeeSchema(name="Shre", id="1" , dept="111")
print(obj_4)
```

## StrictTypes

```python
from pydantic import StrictInt
class EmployeeSchema(BaseModel):
    name:str
    id:StrictInt
    id_1:int

obj = EmployeeSchema(name="Shr",id=2,id_1="1")
print(obj)
```
